    - Es la estructura de control que decide si un bloque de programa se ejecuta.Tambien permiteseleccionar ejecutar un bloque entre un conjunto de opciones.

   - Este tipo de estructura nos permite automatizar la ejecucion repetitiva de codigo mientras ciertas condicoones se cumplan, tenemos 3 variantes de esta estructura de todos los lenjuages de programacion:    
    - Desde... hasta (estructura for)
    - haz mientras (structura while)
    - haz y luego validaz mientras (estructura Do_while)

#Estructura de seleccion de bifurcacion (if-else)

Esta estructura,selecciona una de dos opcioes de ejecuciones, se ejecuta un bloque si la condicion se cumple o se ejecuta otro bloque si la condicion no se cumple(else)

```
if<condicion> :
    <bloque de ejecucuion>
else:    
    <bloque de ejecucuion alterno>



````python
edad = int(input("Dame tu edad"))
print(f"Tecleaste {edad} años")
if edad >= 18:
    print("Eres mayor de edad, ten tu cheve")
else:
    print("Eres menor de edad, ten tu boing de mango")    
print("Fin del programa")
````

````python
edad = int(input("Dame tu edad"))
dinero =float(input("Dae tu cantidad de dinero"))
print(f"Tecleaste {edad} años")
if edad >=18: 
    if dinero >= 36:
     print("Eres mayor de edad,ten tu cheve")
    else:
        print("Eres menor de edad y tienes dinero , ten tu cheve")
else:
    print("Eres menor de edad, ten tu boing")
print("Fin del programa")
````
